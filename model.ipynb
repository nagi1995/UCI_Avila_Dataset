```python
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df_train = pd.read_csv("dataset/avila-tr.txt", sep = ",", header=None, names = ["F1", "F2", "F3", "F4", "F5", "F6", "F7", "F8", "F9", "F10", "target"])
df_train.head()
```

```python
df_train.tail()
```

```python
df_train.info()
```

```python
df_train.describe()
```

```python
df_train["target"].value_counts()
```

```python
df_train["target"] = df_train["target"].astype('category').cat.codes
df_train["target"].value_counts()
```

```python
x_train, y_train = df_train.values[:, :-1], df_train.values[:, -1]
x_train.shape, y_train.shape
```

```python
model = Sequential([Dense(32, input_dim = 10, activation = "relu"),
                    Dense(128, activation = "relu"),
                    Dense(128, activation = "relu"),
                    Dense(32, activation = "relu"),
                    Dense(12, activation = "softmax")])
```

```python
model.compile(optimizer = "adam",
             loss = "sparse_categorical_crossentropy",
             metrics = ["accuracy"])
```

```python
model.fit(x_train, y_train, epochs = 32)
```

```python
df_test = pd.read_csv("dataset/avila-ts.txt", sep = ",", header = None, names = ["F1", "F2", "F3", "F4", "F5", "F6", "F7", "F8", "F9", "F10", "target"])
df_test.head()
```

```python
df_test["target"] = df_test["target"].astype('category').cat.codes
df_test.head()
```

```python
x_test, y_test = df_test.values[:, :-1], df_test.values[:, -1]
x_test.shape, y_test.shape
```

```python
_, accuracy = model.evaluate(x_test, y_test)
```
